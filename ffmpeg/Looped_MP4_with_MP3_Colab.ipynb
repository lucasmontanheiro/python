This notebook:
- Uploads **1 MP4 (video)** and **1 MP3 (audio)**

- Exports a combined MP4 (H.264 + AAC), YouTube-ready

```python

# Replace with your filenames shown above
video_path = 'veovideo.mp4'   # e.g. 'clip.mp4'
audio_path = 'podcast.mp3'   # e.g. 'song.mp3'
output_path = 'output_looped_video.mp4'
```

```python

# Run ffmpeg: loop video until audio ends, keep resolution
!ffmpeg -y -stream_loop -1 -i "$video_path" -i "$audio_path"   -c:v libx264 -preset veryfast -pix_fmt yuv420p   -c:a aac -b:a 128k -shortest "$output_path"
```

```python

# Download the result
from google.colab import files
files.download(output_path)
```
